```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

df = pd.read_csv('fydyc.csv')  # 替换为你的CSV文件路径
df = df.iloc[1:]  # 去除第一行
# 将日期和时间列合并为一个datetime类型的列
df['日期时间'] = pd.to_datetime(df['日期'] + ' ' + df['时间'])
df = df[df['供电预测'].notnull() & df['供电预测']!='y' ]
df = df[df['火电计划'] != '0.0' ]
df.head()
```

```python
df = df[4000:]

plt.figure(figsize=(15,10))  # 设置图形的大小
# 绘制各个列的线
# df['风电预测'] = pd.to_numeric(df['风电预测'])
# df['光伏预测'] = pd.to_numeric(df['光伏预测'])
# df['水电计划'] = pd.to_numeric(df['水电计划'])

df['火电计划'] = pd.to_numeric(df['火电计划'])
df['供电预测'] = pd.to_numeric(df['供电预测'])

# plt.scatter(df['日期时间'], df['风电预测'], label='fdyc',alpha=0.5, s=5)
# plt.scatter(df['日期时间'], df['光伏预测'], label='gfyq',alpha=0.5, s=5)
# plt.scatter(df['日期时间'], df['水电计划'], label='sdjh',alpha=0.5, s=5)

plt.scatter(df['日期时间'], df['火电计划'], label='hdjh',alpha=0.5, s=5)
plt.scatter(df['日期时间'], df['供电预测'], label='gdyc',alpha=0.5, s=5)

# 设置图形的其他元素
plt.xlabel('time')  # x轴标签
plt.ylabel('value')  # y轴标签
plt.title('pred')  # 图形标题
plt.legend()  # 显示图例

# 优化x轴日期显示
plt.gcf().autofmt_xdate()

plt.show()  # 显示图形
```
